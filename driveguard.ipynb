# 4.Real Time Detections

```python
import cv2

# Mencari semua perangkat kamera yang terhubung
for i in range(5):  # Cobalah untuk indeks 0 hingga 4
    cap = cv2.VideoCapture(i)
    if not cap.read()[0]:
        break
    cap.release()

# Menampilkan indeks kamera yang ditemukan
print(f"Kamera USB terdeteksi dengan indeks: {i}")
```

```python

cap = cv2.VideoCapture(1)
while cap.isOpened():
    ret, frame = cap.read()

    #  Make detections
    results = model(frame)

    cv2.imshow('YOLO', np.squeeze(results.render()))

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
# cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Mencari indeks kamera yang terhubung
detected_camera_index = None
for i in range(5):
    cap = cv2.VideoCapture(i)
    if not cap.read()[0]:
        break
    cap.release()
    detected_camera_index = i

# Menampilkan indeks kamera yang ditemukan
if detected_camera_index is not None:
    print(f"Kamera USB terdeteksi dengan indeks: {detected_camera_index}")
    cap = cv2.VideoCapture(detected_camera_index)

    while cap.isOpened():
        ret, frame = cap.read()

        #  Make detections
        results = model(frame)

        cv2.imshow('YOLO', np.squeeze(results.render()))

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
else:
    print("Tidak ada kamera USB yang terdeteksi.")
```

# 5.Train From Scratch

```python
import uuid #Unique Identifier
import os
import time
```

```python
IMAGES_PATH = os.path.join('data', 'images') #/data/images
labels = ['bangun', 'tidur']
number_imgs = 20
```

```python
cap = cv2.VideoCapture(0)
# Loop Through Labels
for label in labels:
    print('Collecting images for {}'.format(labels))
    time.sleep(5)

# Loop Through Images
    for img_num in range(number_imgs):
        print('Collecting images for {}, image number {}'.format(label, img_num))

        # Webcam feed
        ret, frame = cap.read()
        
        # Naming out image path
        imgname = os.path.join(IMAGES_PATH, label+'.'+str(uuid.uuid1()))

        # Writes out image to file
        cv2.imwrite(imgname, frame)

        # Render to the screen
        cv2.imshow('Image Collection', frame)

        # 2 second delay between capture
        time.sleep(2)
        if cv2.waitKey(10) & 0xFF == ord('q'):
             break
cap.release()
cv2.destroyAllWindows()
```

```python
for label in labels:
    print('Collecting images for {}'.format(label))
    for img_num in range(number_imgs):
        print('Collecting images for {}, image number {}'.format(label, img_num))
        imgname = os.path.join(IMAGES_PATH, label+'.'+str(uuid.uuid1()))
        print(imgname)
```
